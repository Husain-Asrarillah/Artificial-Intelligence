```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
# 1. Penyiapan Data
data = {
    'jumlah_baris_kode': [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000],
    'jumlah_bug': [2, 3, 5, 8, 10, 12, 15, 17, 20, 22]
}

df = pd.DataFrame(data)
print(df)
```

```python
x = df[['jumlah_baris_kode']]
y = df['jumlah_bug']
```

```python
# 2. Pemisahan Data (80% Training, 20% Testing)
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
# 3. Pelatihan Model Regresi Linear
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print ("Intercept (a): ", model.intercept_)
print ("Coefficient (b): ", model.coef_[0])
```

```python
# 4. Pengujian Model
y_predic = model.predict(x_test)

print ("Data Testing:")
print (x_test)
print ("Bug Asli:", y_test.values)
print ("Bug Prediksi:", np.round(y_predic, 2))
```

```python
# 5. Evaluasi Error
print ("Nilai Mean Absolute Error (MAE):", metrics.mean_absolute_error(y_test, y_predic))
print ("Nilai Mean Squared Error (MSE):", metrics.mean_squared_error(y_test, y_predic))
print ("Nilai Root Mean Squared Error (RMSE):", np.sqrt(metrics.mean_squared_error(y_test, y_predic)))
```

```python
# 6. Prediksi Data Baru (Diperbaiki agar tidak memunculkan warning)
data_baru = pd.DataFrame({'jumlah_baris_kode': [1200]})
prediksi = model.predict(data_baru)
print("\nPrediksi Bug untuk 1200 Baris Kode:", round(prediksi[0]))
```

```python
# 7. Visualisasi
plt.scatter(x,y, label="Data Asli")
plt.plot(x, model.predict(x), color="red", label="Garis Regresi")
plt.xlabel("Jumlah Baris Kode")
plt.ylabel("Jumlah Bug")
plt.title("Prediksi Jumlah Bug berdasarkan Jumlah Baris Kode")
plt.legend()
plt.show()
```

**Analisis Hasil Praktikum Regresi Linear**

**1. Apa yang dipelajari dari model ini?**
Setelah saya menjalankan kodingan di atas, model berhasil menemukan hubungan antara jumlah baris kode dengan jumlah bug.
*   Nilai Koefisien (Slope): Nilainya positif (sekitar 0.022). Artinya, ada hubungan yang searah: kalau baris kodenya makin banyak, maka jumlah bug-nya juga bakal makin banyak. Sederhananya, setiap kita nambah 100 baris kode, kemungkinan bakal muncul sekitar 2 bug baru.
*   Nilai Intercept: Ini adalah titik awal. Karena nilainya mendekati nol, ini masuk akal secara logika. Kalau kita belum nulis kode sama sekali (0 baris), ya harusnya nggak ada bug yang muncul.


**2. Seberapa akurat modelnya?**
Untuk melihat model ini "pintar" atau tidak dalam menebak, saya melihat nilai error-nya (MAE dan RMSE):

* Nilai MAE-nya cukup kecil (di bawah 1). Ini artinya, tebakan model saya rata-rata hanya meleset sedikit sekali dari data aslinya.
* Kalau dilihat dari grafik scatter plot, titik-titik data asli menempel sangat dekat dengan garis merah (garis regresi). Ini membuktikan kalau model regresi linear ini sangat cocok untuk data yang saya buat.

**3. Mencoba prediksi data baru**
Saya mencoba mengetes model ini untuk memprediksi proyek yang lebih besar, yaitu 1200 baris kode. Hasilnya, model menebak akan ada sekitar 26 bug di dalam kode tersebut. Ini membantu saya untuk bersiap-siap melakukan testing lebih ketat kalau kodingannya makin panjang.
